```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# **Line Plot**

**1.Simple Axis**

```python
x=[0,1,2,3,4,5]
y=[0,15,20,35,48,50]
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**2.Condition in axis**

```python
x=[0,1,2,3,4,5]
y=[i**3 for i in x]
plt.plot(x,y)
plt.title("Condition line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**3.Default Axis**

```python
plt.plot([10,20,30,40,50])  #values will be taken as y axis by default
plt.show()
```

# Markers

```python
x=np.arange(0,10,2)
y=[i**2 for i in x]
plt.title("Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,marker='o',mec='k',mfc='m',ms=20)#mec=marker edge colour, mfc=marker face colour,ms=marker size
plt.show()
```

# Lines(Style,width,colour)

```python
x=np.arange(0,10,2)
y=x**2
y1=x*2
print(x)
print(y)
plt.plot(x,y,ls="dashed",lw=10,c="r")
plt.plot(x,y1,marker='p',mec='b',mfc='k')#ls=linespace,lw=linewidth,c=lineccolor
plt.show()
```

# Font Properties

```python
x=np.arange(0,20,3)
y=x*10
plt.plot(x,y)
a1={'font':'ALGERIAN','color':'red','size':20}
a2={'font':'calibri','color':'green','size':15}
a3={'font':'Times New Roman','color':'green','size':15}
plt.title("Checking Font Properties",fontdict=a1)
plt.xlabel("Values1",fontdict=a2)
plt.ylabel("Values2",fontdict=a3)
plt.show()
```

# GRID FUNCTION WITH LINE PROPERTIES

```python
x=[1,2,3,4,5,6]
y=[20,60,10,90,50,30]
plt.plot(x,y)
plt.grid(axis='both',ls='dashdot',c='r')
plt.show()
```

# Subplots

```python
x=[1,2,3,4,5,6]
y1=[10,52,34,49,59,26]
y2=[31,25,83,48,55,60]
y3=[21,25,73,28,35,20]
y4=[21,85,23,78,75,10]
plt.suptitle("Multiple Plots")

plt.subplot(2,2,1)
plt.title("Plot-1")
plt.plot(x,y1)


plt.subplot(2,2,2)
plt.title("Plot-2")
plt.plot(x,y2)


plt.subplot(2,2,3)
plt.title("Plot-3")
plt.plot(x,y3)


plt.subplot(2,2,4)
plt.title("Plot-4")
plt.plot(x,y4)


plt.show()
```

# Legends

**1.Without Legends function,labels cannot be displayed**

```python
x=np.arange(0,5,1)
y1=x**2
y2=x**3
plt.plot(x,y1,label="Square")
plt.plot(x,y2,label="Cube")
plt.show()
```

```python
x=np.array([0,1,2,3,4,5])
y1=x**2
y2=x**3
plt.plot(x,y1,label="Square")
plt.plot(x,y2,label="Cube")
plt.legend() #Now labels will be displayed
plt.show()
```

**2.Another way to show Legends**

```python
x=np.array([0,1,2,3,4,5])
y1=x**2
y2=x**3
plt.plot(x,y1,x,y2)
plt.legend(["Square","Cube"])
plt.show()
```

**3.Location Of Legends**

```python
x=np.array([0,1,2,3,4,5])
y1=x**2
y2=x**3
plt.plot(x,y1,x,y2)
plt.legend(["Square","Cube"],loc="upper right")
plt.show()
```

**3.Transparency of legends box**

```python
x=np.array([0,1,2,3,4,5])
y1=x**2
y2=x**3
plt.plot(x,y1,x,y2)
plt.legend(["Square","Cube"],loc="upper left",framealpha=0.6)
plt.show()
```

**4.legends box colour**

```python
x=np.array([0,1,2,3,4,5])
y1=x**2
y2=x**3
plt.plot(x,y1,x,y2)
plt.legend(["Square","Cube"],loc="upper left",framealpha=0.6,facecolor="green")
plt.show()
```

**5.legends box edge colour**

```python
x=np.array([0,1,2,3,4,5])
y1=x**2
y2=x**3
plt.plot(x,y1,x,y2)
plt.legend(["Square","Cube"],loc="upper left",framealpha=0.6,facecolor="green",edgecolor="black")
plt.show()
```

**6.Shadow & Fancybox & Font size**

```python
x=np.array([0,1,2,3,4,5])
y1=x**2
y2=x**3
plt.plot(x,y1,x,y2)
plt.legend(["Square","Cube"],loc="upper left",framealpha=0.6,facecolor="green",edgecolor="black",fancybox=True,shadow=True,fontsize=15)
plt.show()
```

# Scatter Plot

```python
import random
x=[1,2,3,4,5,6,7,8]
y=[10,20,30,40,50,60,70,80]
random.shuffle(y)
print(x,y)
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y1=[10,20,30,40,50,60,70,80]
y2=[10,20,30,40,50,60,70,80]

random.shuffle(y1)
print(x,y1)
plt.scatter(x,y1,color="black",marker="s",s=100,linewidth=2,alpha=0.5)

random.shuffle(y2)
print(x,y2)
plt.scatter(x,y2,color="green",marker="^",s=100)


plt.show()
```

# Stacked Bar Chart

**1.Overlapping Stacked Barchart**

```python
years=["2017","2018","2019","2020"]

cse=[75,80,47,82]
mech=[46,58,35,52]
ece=[67,59,72,62]


plt.bar(years,cse,width=0.4)
plt.bar(years,mech,width=0.4)
plt.bar(years,ece,width=0.4)

plt.ylim(0,130)

plt.show()
```

**2.Non-Overlapping Stacked Barchart**

```python
years=["2017","2018","2019","2020"]

cse=[75,80,47,82]
mech=[46,58,35,52]
ece=[67,59,72,62]

ece_start=[cse[i]+mech[i] for i in range (len(cse))]

plt.bar(years,cse,width=0.4,color="black",label="cse")
plt.bar(years,mech,bottom=cse,width=0.4,color="green",label="mech")
plt.bar(years,ece,bottom=ece_start,width=0.4,color="red",label="ece")

plt.ylim(0,300)
plt.title("Placement Comparison")
plt.legend()
plt.show()
```

```python
years=["2017","2018","2019","2020"]

cse=[75,80,47,82]
mech=[46,58,35,52]
ece=[67,59,72,62]
w=0.2
plt.figure(figsize=(10,5))

#Locations to Plot
cse_bar=np.arange(len(years))
print(cse_bar)
mech_bar=[i+w for i in cse_bar] 
print(mech_bar)
ece_bar=[i+w for i in mech_bar] 
print(ece_bar)

plt.bar(cse_bar,cse,width=w,label="cse")
plt.bar(mech_bar,mech,width=w,label="mech")
plt.bar(ece_bar,ece,width=w,label="mech")

plt.legend()
plt.title("Placement Comparison")
plt.ylim(0,120)
plt.xticks(cse_bar+w,years)
plt.xlabel("Years")
plt.ylabel("Placement Count")

plt.show()
```

# Piechart

```python
student_performance=["Excellent","Good","Average","Satisfactory"]
students=[20,30,15,8]
plt.figure(figsize=(8,10))
plt.pie(students,labels=student_performance,startangle=90,explode=[0.5,0,0.2,0],shadow=True,colors=["grey","orange","pink","blue"],autopct="%2.1f%%")
plt.legend(title="Performances")
plt.show()
```
