# Conditional Changes

```python
df_sales=pd.read_csv("F:\\Data Science\\vgsales.csv")
```

```python
df=pd.DataFrame(df_sales)
```

```python
df
```

```python
df["Percentage_NA"]=(df["NA_Sales"]/df["Global_Sales"])*100
df
```

```python
df["Grade_NA"]=None
df
```

```python
df.loc[df["Percentage_NA"]>50,["Grade_NA"]]="A"
```

```python
df.loc[(df["Percentage_NA"]>30) & (df["Percentage_NA"]<=50),["Grade_NA"]]="B"
```

```python
df.loc[df["Percentage_NA"]<30,["Grade_NA"]]="C"
```
